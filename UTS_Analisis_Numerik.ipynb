```python
print("\033c")
import numpy as np
import matplotlib.pyplot as plt
import time

# Mencari Akar Persamaan Polinomial
#a = 0.2; b = 1.4; c = -8; d = -23

# Masukkan Nilai Variabel (Input)
dx = 0.0001
err_m = 0.001
x_awal_1 = -10; x_akhir_1  = -9
x_awal_2 = -2.5; x_akhir_2 = -1.5
x_awal_3 = 5.0; x_akhir_3  = 6.0

# Mencari Akar Pertama
x = x_awal_1
y = 0.2*x**3 + 1.4*x**2 + -8*x -23

i = 0
while(x < x_akhir_1 and abs(y) > abs(err_m)):
    i = i + i
    x = x + dx
    y = 0.2*x**3 + 1.4*x**2 + -8*x -23
    print("Iterasi ke-", i)
    print("x =", x)
    print("y =", y)

# Mencari Akar Kedua
x = x_awal_2
y = 0.2*x**3 + 1.4*x**2 + -8*x -23

i = 0
while(x < x_akhir_2 and abs(y) > abs(err_m)):
    i = i + i
    x = x + dx
    y = 0.2*x**3 + 1.4*x**2 + -8*x -23
    print("Iterasi ke-", i)
    print("x =", x)
    print("y =", y)

# Mencari Akar Ketiga
x = x_awal_3
y = 0.2*x**3 + 1.4*x**2 + -8*x -23

i = 0
while(x < x_akhir_3 and abs(y) > abs(err_m)):
    i = i + i
    x = x + dx
    y = 0.2*x**3 + 1.4*x**2 + -8*x -23
    print("Iterasi ke-", i)
    print("x =", x)
    print("y =", y)

y = abs(y)
print(" ")
print("VARIABEL (INPUT)")
print("dx           =", dx)
print("error margin =", err_m)
print("x_awal_1     =", x_awal_1)
print("x_akhir_1    =", x_akhir_1)
print("x_awal_2     =", x_awal_2)
print("x_akhir_2    =", x_akhir_2)
print("x_awal_3     =", x_awal_3)
print("x_akhir_3    =", x_akhir_3)

print("HASIL AKHIR")
if(y<err_m):
    print("Pencarian akar berhasil.")
else:
    print("Pencarian akar gagal, silahkan perkecil dx atau perbesar error margin.")
print("Banyaknya iterasi:", i)
print("x =", x)
print("y =", y)
print(" ")

# -----------------------------------------------Visualisasikan Kurva-----------------------------------------------------------
x = np.linspace(-2, 3, 10000)
plt.figure(figsize=(6,6.5))
sumbu_x   = x -x -0
y         = 2*x**3 - 3*x**2 - 5*x + 3
plt.plot(x,   sumbu_x,             '-k')
plt.plot(x,   (0.01-x**2)**0.5,    '-k')
plt.plot(x,   -((0.01-x**2)**0.5), '-k')
plt.plot(x,   y,                   '-g', label="y")
plt.legend(loc='upper left')
plt.grid()
plt.show()
```
